#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> [used to visualize the frequency of values in numerical columns // used to visualize value-counts seperated in bins]
- Density Plot / KDE
> [Like a histogram it is used to visualize the frequency of values but it creates a smooth continuous curve]
- Bar Chart
> [used to visually seperate the data based on its category and show the values using lines]  ____(categorical data)____
- Line Plot
> [show the data by joining the values through connecting them using lines]
- Univariate scatter plot
> [This plots different observations/values of the same variable corresponding to the index/observation number.]
- Violin Plot
> [similar to the box plot, a violin plot shows the distribution of the data and it includes a rotated KDE curve]
- Box Plot
> [also known as a wisker plot is used to show the distribution of data and helps with the detection of outliers]
- Pie Chart
> [used to visually seperate the data based on its category and show the values through differing slice sizes]  ____(categorical data)____
- Rug Plot
> [it's a way to show the distribution of data using short ticks. The density of distribtion is illustrated by the density of the ticks]

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt 

# set standard plot type as seaborn plot
sns.set()

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
```

```python
df.age.value_counts(bins=8, sort=False)
```

```python
plt.figure(figsize=(12,6))
sns.set_theme(palette='turbo',font_scale=1.3)
sns.histplot(data=df.age,bins=8,kde=True)
# sets x and y labels
plt.xlabel("Age", fontsize=14)
plt.ylabel("Number of Managers", fontsize=14)

# sets title
plt.title("Distribution of Manager Ages in Survey", fontsize=15);
```

##### Insights:
[As the age gets higher, the higher chance a person has of being a manager.]

### Plot for 'industry' column

```python
# create plot for industry column
```

```python
plt.figure(figsize=(12,6))
data=df.industry.value_counts()
labels=df.industry.unique()
colors=['r','b','y','g']
figure=plt.pie(data,explode=[0.01,0.01,0.01,0.01],labels=labels,autopct='%.0f%%',colors=colors,shadow=True)
plt.title('Percentage of Industries in Survey',fontsize=16)
plt.tight_layout()
```

##### Insights:
[data gives equal representaion to all industries ]

### Plot for 'salary' column

```python
# create plot for salary column
```

```python
plt.figure(figsize=(12,6))
sns.set_theme(palette='coolwarm',font_scale=1.2)
figure=sns.histplot(df.salary,kde=True,bins=10)
plt.title('Salaries of Managers',fontsize=18)
plt.xlabel('Salary',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
df.salary.mean()
```

##### Insights:

[Salary is a bell curve centered at the mean which equals 101416]

### Plot for 'industry_experience' column

```python
# create plot for industry experience column
```

```python
plt.figure(figsize=(12,6))
figure=sns.violinplot(df.industry_experience,color='#ADD8E6')
plt.title('Industry Experience', fontsize=18)
plt.xlabel('Experience in Years', fontsize=15)
df.industry_experience.mean()
```

##### Insights:
[The average experience is 18 years]

### Plot for 'education' column

```python
# create plot for education levels
```

```python
df.education.value_counts()
```

```python
plt.figure(figsize=(12,6))
data=df.education.value_counts()
labels=["Bachelor's degree","Master's degree", 'Some College','High School']
plt.pie(data,labels=labels,explode=[0,0.01,0,0.2],autopct='%.0f%%',shadow=True,pctdistance=.5)
plt.title("Manager's Education Levels",fontsize=18)
plt.tight_layout()
```

##### Insights:

[Most managers had a degree. Most managers had a Bachelor's followed by Master's.]

### Plot for 'gender' column

```python
# create plot for gender column
```

```python
plt.figure(figsize=(12,6))
data=df.gender.value_counts()
sns.set_theme(font_scale=1.2)
sns.countplot(df.gender,palette='viridis',)
plt.title('Distribution based on Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
```

##### Insights:

[data is split equally between genders]

## Part 2 - Summary Statistics

### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
> The mean is the average of the data
- median
> The median is the middle score for a set of data that has been arranged in order of magnitude. The median is less affected by outliers and skewed data.
- mode
> The mode is the most frequent score in our data set. On a histogram it represents the highest bar in a bar chart or histogram. You can, therefore, sometimes consider the mode as being the most popular option.
- central tendency
> the statistical measure that identifies a single value as representative of an entire distribution
- distribution
> the way in which something is shared out among a group or spread over an area
- variance
> a measure of dispersion that takes into account the spread of all data points in a data set. Also called mean square distance
- standard deviation
> measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance

###### Reflection: How did mean and median differ above? Why do you think that is?
