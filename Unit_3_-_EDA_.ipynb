### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
> The mean is the average of the data
- median
> The median is the middle score for a set of data that has been arranged in order of magnitude. The median is less affected by outliers and skewed data.
- mode
> The mode is the most frequent score in our data set. On a histogram it represents the highest bar in a bar chart or histogram. You can, therefore, sometimes consider the mode as being the most popular option.
- central tendency
> the statistical measure that identifies a single value as representative of an entire distribution. Measures of central tendency are mean,median, mode.
- distribution
> the way in which something is shared out among a group or spread over an area
- variance
> a measure of dispersion of data points around the mean value. Also called mean square distance.

- standard deviation
> measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance

### What is the correlation between salary and education?

#### Your Hypothesis = [The higher the education the higher the salary. Positive correlation]

```python
# calculate correlation and pvalue of salary and education with spearmanr method
```

```python

spearmanr(df['salary'], df['education'])
```

[the p-value is approximatly zero, positive correlation ]

#### Are these results statistically significant? How do you know?

[yes, since the p-value is approximatly zero which is less than our significance threshold of 0.05 we can reject the null hypothesis]

### What is the correlation between salary and age?

#### Your Hypothesis [The older a person is the higher the salary]

```python
# calculate correlation and pvalue of salary and age with pearsonr method
```

```python
pearsonr(df['salary'], df['age'])
```

[the p-value is zero, positive correlation]

#### Are these results statistically significant? How do you know? 

[yes, since the p-value is zero which is less than our significance threshold of 0.05 we can reject the null hypothesis]

### What is the correlation between salary and industry experience? 

#### Your Hypothesis = [the more experience the higher the salary]

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
```

```python

pearsonr(df['salary'], df['industry_experience'])
```

[the p-value is approximatly zero, positive correlation]

#### Are these results statistically significant? How do you know?

[yes, since the p-value is zero which is less than our significance threshold of 0.05 we can reject the null hypothesis]

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
```

```python
cm=pd.crosstab(index=df_num.industry,columns=df_num.education,values=df_num.salary, aggfunc=np.mean).round()
cm
```

```python
# save above result to new variable
```

```python

plt.figure(figsize=(16,10))

ax = sns.heatmap(cm, annot=True, linewidths=.5, cmap='Spectral', fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
```
